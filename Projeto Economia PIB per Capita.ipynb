```python
# Bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lendo a base de Dados
Base_Dados = pd.read_excel('/content/Dados_Pib.xlsx')

# Verificando os Dados
Base_Dados.head()
```

```python
Base_Dados.groupby( by=['Territorialidades', 'Ano']).mean()
```

```python
# Sistemas de Grids

# Cor de Fundo
Cor_Fundo = '#f5f5f5'

# Criar o sistema de Grids
Grid_Graficos = sns.FacetGrid(Base_Dados, col='Territorialidades', hue='Territorialidades', col_wrap=4)

# Adicionar gráfico linha em cada gráfico
Grid_Graficos = Grid_Graficos.map(plt.plot, 'Ano', 'PIB per capita')

# Adicionar sombra + Ajuste de titulo
Grid_Graficos = Grid_Graficos.map(plt.fill_between, 'Ano', 'PIB per capita', alpha=0.2).set_titles('{col_name} Territorialidades')

# Filtrar o titulo
Grid_Graficos = Grid_Graficos.set_titles('{col_name}')

# Adicionar um subtitulo
Grid_Graficos = Grid_Graficos.fig.suptitle('Evolução da Renda per capita por Estado \n Esse relatório foi elaborado  no treinado "Python para Análise de Dados" \n by= Joao Nobrega', fontsize=18)

# Ajustando
plt.subplots_adjust(top=0.92)
```
