```python
# Libs Necess치rias

# Libs para Modelagem e Matrizez
import numpy as np
import pandas as pd

# Libs para ana치lises gr치ficas
import matplotlib.pyplot as plt
import seaborn as sns

# Lib para ignorar avisos
import warnings

# Desabilitando avisos
warnings.filterwarnings('ignore')
```

```python
Base_Dados = pd.read_csv('/content/Dados_Indendio.csv', encoding='latin-1')
Base_Dados.head()
```

```python
# Verificando campos nulos
Base_Dados.isnull().sum()
```

```python
# Nulos plot
plt.figure(figsize=(14, 5))
plt.title('An치lise de campos nulos')
sns.heatmap(Base_Dados.isnull(), cbar=False);
```

```python
# Estatisticas
Base_Dados.describe()
```

```python
# Informa칞칫es
Base_Dados.info()
```

```python
# Campos unicos
Base_Dados.nunique()
```

```python
# Analise por ano dos incendios
Analise = Base_Dados.groupby(by=['year']).sum().reset_index()
Analise.head()

# Tamanho
plt.figure(figsize=(12, 5))

# Estilo do Gr치fico
plt.style.use('ggplot')

# Gr치fico
plt.title('Total inc칡ndios no Brasil 1997 - 2017', loc='left', fontsize=14)
sns.lineplot(data=Analise, x='year', y='number', estimator='sum', lw=2, color='#ff5555', alpha=0.85)

# Labels
plt.xlabel('Quantidade')
plt.ylabel('Per칤odo');
```

```python
# Analise por ano dos incendios
Analise_02 = Base_Dados.groupby(by=['year','month']).sum().reset_index()
Analise_02.head()

# Tamanho
plt.figure(figsize=(12, 5))

# Gr치fico
plt.title('Inc칡ndios por m칡s', loc='left', fontsize=14)

sns.boxplot(data=Analise_02, x='month', y='number', palette='coolwarm', saturation=1, width=0.9, linewidth=2,
            order=['Janeiro', 'Fevereiro', 'Mar칞o', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'])

# Labels
plt.xlabel('M칡s')
plt.ylabel('N칰mero de inc칡ncdios');
```

```python

# Analise por ano dos incendios
Analise_03 = Base_Dados.groupby(by=['state']).sum().reset_index()[['state','number']].sort_values('number', ascending=False)
Analise_03.head()

# Tamanho
plt.figure(figsize=(12, 5))

# Gr치fico
plt.title('Estados com maior n칰mero de inc칡ndios', loc='left', fontsize=14)

#Gr치fico
plt.bar(Analise_03.state, Analise_03['number'], color='#f44e3f')

# Labels
plt.ylabel('Quantidade')
plt.xticks(rotation=90);
```

```python
# Estados TOP 10
Lista_TOP10 = Analise_03['state'][0:10].values

# Tamanho
plt.figure(figsize=(12, 5))

# Gr치fico
plt.title('Top 10 Estados com inc칡ndios', loc='left', fontsize=14)

for Coluna in Lista_TOP10:

  # Filtrar o estado
  Filtro = Base_Dados.loc[Base_Dados['state'] == Coluna]

  # Agrupar os valores para sumarizar
  Analise = Filtro.groupby(by=['year']).sum().reset_index()

  # Plot
  sns.lineplot(data=Analise, x='year', y='number', lw=2, alpha=0.9)

  # Labels
plt.xlabel('Per칤odo')
plt.ylabel('Quantidade')

# Legenda
plt.legend(Lista_TOP10, bbox_to_anchor=(1, 0.7));
```
