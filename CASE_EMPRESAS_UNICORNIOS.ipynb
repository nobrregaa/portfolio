```python
#importar as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
#Ler os dados
Base_Dados = pd.read_csv("/content/Startups+in+2021+end.csv")

#Verificar Dimensão
Base_Dados.shape
```

```python
# Primeiros registros
Base_Dados.head()
```

```python
# Verificando Colunas
Base_Dados.columns
```

```python
# Renomear Colunas
Base_Dados.rename( columns={
    'Unnamed: 0':'ID',
    'Company':'Empresa',
    'Valuation ($B)':'Valor($)',
    'Date Joined':'Data_Abertura',
    'Industry':'Setor',
    'Country':'Pais',
    'Select Investors':'Investidores',
    'City':'Cidade',

}, inplace=True)
```

```python
# Verificar Tipo da informação
Base_Dados.info()
```

```python
# Grafico Analise dos Setores
plt.figure(figsize=(15,6))
plt.title('Analise dos Setores')
plt.bar(Base_Dados['Setor'].value_counts().index, Base_Dados['Setor'].value_counts());
plt.xticks(rotation=45, ha='right');
```

```python
Analise = round(Base_Dados['Pais'].value_counts( normalize=True) * 100, 1)
```

```python
# Plot geral dos Países
plt.figure(figsize=(15,6))
plt.title('Analise dos Países')
plt.pie(Analise,
        labels = Analise.index,
        shadow=True,
        startangle=90,
        autopct='%1.1f%%');
```

```python
# Plot geral dos Países
plt.figure(figsize=(15,6))
plt.title('Analise dos Países - Top 10')
plt.pie(Analise.head(10),
        labels = Analise.index[0:10],
        shadow=True,
        startangle=90,
        autopct='%1.1f%%');
```

```python
# Conversão para Data
Base_Dados['Data_Abertura'] = pd.to_datetime(Base_Dados['Data_Abertura'])
Base_Dados['Data_Abertura'].head()
```

```python
# Extrair o Ano e Mês
Base_Dados['Ano'] = Base_Dados['Data_Abertura'].dt.year
Base_Dados['Mes'] = Base_Dados['Data_Abertura'].dt.month
Base_Dados.head()
```

```python
# Tabela Analítica
Analise_Agrupada = Base_Dados.groupby(by=['Pais',  'Ano', 'Mes', 'Empresa']).count()['ID'].reset_index()

Analise_Agrupada
```

```python
Analise_Agrupada.loc[
    (Analise_Agrupada['Pais'] == 'Brazil')
]
```

```python
# Transformando a coluna Valor
Base_Dados['Valor($)'] = pd.to_numeric(Base_Dados['Valor($)'].apply(lambda Linha: Linha.replace('$', '')))
Base_Dados.head()
```

```python
# Tabela Analítica
Analise_Pais = Base_Dados.groupby(by=['Pais'])['Valor($)'].sum().reset_index()

Analise_Pais.head()
```

```python
Analise_Valor = Analise_Pais.sort_values(by='Valor($)', ascending=False)
Analise_Valor.head()
```

```python
plt.figure(figsize=(15,6))
plt.title('Analise do Valor por País')
plt.plot(Analise_Valor['Pais'], Analise_Valor['Valor($)'])
plt.xticks(rotation=45, ha='right');
```
